```python
# Colour Transparency

# RGBA - Red, Green, Blue, Alpha (transparency)
```

```python
red = Image.open("red_color.jpg")
```

```python
red.putalpha(125)
```

```python
red
```

```python
blue = Image.open("blue_color.png")
```

```python
blue.putalpha(128) #completely transparent
```

```python
red.paste(blue, box=(0,0), mask=red)
```

```python
red
```

```python
red.save("transparent.png")
```
